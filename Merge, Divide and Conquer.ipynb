```python
import pandas as pd
import numpy as np
```

```python
# Read in the files
# Was originally planned to be the main input/output file

#p_2017 = pd.read_csv('../Data/cleaned/parking_2017_cleaned.csv')
p_geom = pd.read_csv('../Data/cleaned/bay_geom_cleaned.csv')
p_sensor = pd.read_csv('../Data/cleaned/parking_sensors_cleaned.csv')
p_bays = pd.read_csv('../Data/cleaned/bays_2017.csv')
```

## Merging Geometry onto Bay Aggregates

```python
p_bays.shape
```

```python
# Merging geometry table and sensor table left onto aggregated bays 

tableau = p_bays.merge(p_sensor, left_on=['street_marker'],\
                                right_on=['st_marker_id'], how='left').\
                                merge(p_geom, left_on='street_marker',\
                                right_on='marker_id', how='left')
tableau.drop(columns=['st_marker_id','bay_id_x','duration_allowed_y',\
                     'duration_allowed', 'Unnamed: 0', 'marker_id', 'bay_id_y'], inplace=True)

tableau.rename(columns = {'duration_allowed_x':'duration_allowed',
                         'avg_perc_empty':'avg_empty_perc'}, inplace=True)

# Changing units of duration allowed to min, and type to int

tableau['duration_allowed'] = tableau['duration_allowed'].div(60).astype('int')

tableau.head()
```
